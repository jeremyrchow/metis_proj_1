```python
import pandas as pd
```

```python
df=pd.read_csv('test_data.csv', header=0, names = ["ca","remote_unit",\
               "scp_address","station","line_name","division","date","time",\
               "description","entry_cumulative","exit_cumulative"])
df=df.drop(columns = ['ca','remote_unit','division'])
```

```python
df['date'] = pd.to_datetime(df['date'],format = '%m/%d/%Y')
df['time'] = pd.to_datetime(df['time'],format = '%H:%M:%S')
df.head()
```

```python
df2=df.groupby(['station','scp_address','date']).apply(pd.DataFrame.sort_values,'time')\
df2
```
